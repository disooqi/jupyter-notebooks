```python
import os
import codecs
```

```python
i=0
infile=""
outfile=""
scheme=""
args_flag = 0

# os.getenv is equivalent, and can also give a default value instead of `None`
dataDirectory = os.getenv('FarasaDataDir', None)

if not dataDirectory:
    print('"FarasaDataDir" Environment variable is not set')

dataDirectory
```

```python
print("Initializing the system ....")
```

```python
roots_path = os.path.join(dataDirectory, 'roots.txt')
hmRoot = dict()
with codecs.open(roots_path, encoding='utf-8') as roots:
    for root_line in roots:
        tl = root_line.strip().split()
        if len(tl) == 2:
            hmRoot[tl[0]] = float(tl[1])
        else:
            print('some problem here')

hmRoot
```

```python
template_path = os.path.join(dataDirectory, 'template-count.txt')
hmTemplate = dict()
with codecs.open(template_path, encoding='utf-8') as template:
    for template_line in template:
        tl = template_line.strip().split()
        if len(tl) == 2:
            hmTemplate[tl[0]] = float(tl[1])
        else:
            print('some problem here')

hmTemplate
```
